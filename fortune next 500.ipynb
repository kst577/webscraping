### Extracting company names

```python
names=soup.findAll('a',href=True)

#to extract the company names
```

```python
company=[]
for i in names:
    company.append(i.text)
company_names=company[52:]
```

```python
updated=[]
for i in company_names:
    updated.append(i.replace(' ','-'))
```

```python
updated_company=[]
for i in updated:
    updated_company.append(i[1:])
```
